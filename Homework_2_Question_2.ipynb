```python
import requests
import json
import pandas as pd

#API Key
key = '27208c853ea24aa192311823243103'

#Declare cities
cities = ['Washington', 'Boston', 'Houston', 'San Antonio', 'Phoenix', 'Denver', 'San Francisco', 'Philadelphia', 'New York', 'Los Angeles']

#Lists for the weather information
temperature = []
humidity = []
wind_speed = []
uv = []
description = []

for city in cities:
  #API endpoint
  url = f'http://api.weatherapi.com/v1/current.json?key={key}&q={city}'

  r = requests.get(url) # Request the data from the URL
  if(r.status_code == 200):
    data = r.json() # Extract the JSON to data variable

    #Append weather data into list
    temperature.append(data['current']['temp_f']) #Appending temperature data
    humidity.append(data['current']['humidity']) #Appending humidity data
    wind_speed.append(data['current']['wind_mph']) #Appending minimum temperature data
    uv.append(data['current']['uv']) #Appending maximum temperature data
    description.append(data['current']['condition']['text']) #Appending weather description data
  else:
    print("API request was unsuccessful.")

#Creating a series
weather_info = {
    'City' : cities,
    "Temperature" : temperature,
    "Wind Speed" : wind_speed,
    "UV Index" : uv,
    "Humidity" : humidity,
    'Weather Description' : description
}

#Creating the dataframe
df = pd.DataFrame(weather_info)
df #Printing the dataframe
```

```python
df.to_csv("Weather Data.csv", index=False) #Creating CSV file
df.describe() #Printing descriptive statistics
```
