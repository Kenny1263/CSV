```python
#Function to request the original date
def readoriginaldate():
  date = input("Enter a date (Month/Day/Year): ") #Prompt user to input the date
  print(f"You Entered {date}") #Printing the date entered

  return date #returning the date entered

#Function to split the original date
def breakoriginaldate(date): #Recieves the data parameter
  parts = date.split('/') #Split the date at the "/"

  month = parts[0] #The first split which represents the month
  day = parts[1] #The second split which represents the day
  year = parts[2] #The third split which represents the year

  print(f"{date} is the original date") #Prints the original date
  print(f"{month} is the month   {day} is the day   {year} is the year") #Prints the parts of the date

  return month, day, year #Returns the month, day and year

#Function to print the date in three different ways
def printdate3ways(day, month, year): #Recieves the day, month and year parameters
    #European Format
    european_format = f"{day}-{month}-{year}" #Creating the European format variable
    print(f"{european_format} is the European way of printing")  #Print the date in the European way

    #American Format
    #Define a dictionary mapping month numbers to their English names
    month_names = {
        '1': 'January', '2': 'February', '3': 'March', '4': 'April',
        '5': 'May', '6': 'June', '7': 'July', '8': 'August',
        '9': 'September', '10': 'October', '11': 'November', '12': 'December'
    }

    #Assign the respective englsh name to the month
    english_month = month_names[month]

    american_format = f"{english_month} {day}, {20}{year}" #Creating the American format variable
    print(f"{american_format} is the American way of printing") #Print the date in the American way

    #Full Format
    if int(month) < 10: #Adds a 0 to the beginning of the value if the month is lower than 10
      month = f'0{month}'
    if int(day) < 10: #Adds a 0 to the beginning of the value if the day is lower than 10
      day = f'0{day}'

    full_format = f"{month}-{day}-{year}" #Creating the full format variable
    print(f"{full_format} is the full way of printing") #Print the date in the full way


#Main Function
def main():
  dates = [] #Creates a list for the dates to be entered
  while True:
    dates.append(readoriginaldate()) #Appends the entered dates into the dates list
    cont = input("Continue entering more dates? (Y/N): ") #Prompts user to continue or stop entering dates
    if cont == "N" or cont == "n":
      print('\n') #Skipping line for aesthetic purposes
      break
    else:
      print() #Skipping line for aesthetic purposes


  #Function to iterate through all the dates in the dates list
  for date in dates:
    month, day, year = breakoriginaldate(date) #Calls the breakoriginaldate function
    printdate3ways(day, month, year) #Calls the printdate3ways function
    print('\n\n\n') #Skipping lines



#Calling the main function
main()
```
